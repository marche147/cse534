## Setup P4 docker

```python
threads = {}
#targets = [d1_name, d2_name, d3_name, d4_name, attacker_name]
targets = [s1_name, s2_name, d1_name, d2_name, d3_name, d4_name, attacker_name]

try:
    for name in targets:
        print("Setting up docker for node {}".format(name))
        node = slice.get_node(name=name)
        if type(ip_address(node.get_management_ip())) is IPv4Address:
            management_ip_type = "IPv4"
        else:
            management_ip_type = "IPv6"
        node.upload_file("./setup_p4.sh", "./setup_p4.sh")
        tnode = node.execute_thread(f"chmod +x ./setup_p4.sh && sudo sh -c './setup_p4.sh {management_ip_type}'")
        threads[name] = tnode
except Exception as e:
    traceback.print_exc()
```

```python
try:
    for node, thread in threads.items():
        stdout, stderr = thread.result()
        print(f"Config thread node {node} complete")
        #print(f"stdout: {stdout}")
        #print(f"stderr: {stderr}")
except Exception as e:
    print(f"Error: {e}")
    traceback.print_exc() 
```

## Setup TAP virtual network

```python
try:
    for host in [d1_name, d2_name, d3_name, d4_name, attacker_name, s1_name, s2_name]:
        print("Download l2direct.py on {}".format(host))
        node = slice.get_node(name=host)
        node.execute_thread("rm -rf l2direct.* && wget https://gist.githubusercontent.com/marche147/9ab402a694d27f3330ac70b1cecc2067/raw/b81f7390185ced82701d796b00a3db7a4eeb8621/l2direct.py")
except Exception as e:
    traceback.print_exc()
```

```python
tap_networks = {
    d1_name: ['undefined', s1_addr, 5001],
    d2_name: ['undefined', s1_addr, 5002],
    d3_name: ['10.0.1.10/24', s2_addr, 5001],
    d4_name: ['10.0.1.11/24', s2_addr, 5002],
    attacker_name: ['undefined', s1_addr, 5003],
    s1_name: ['undefined', s2_addr, 5004],
}

s1_node = slice.get_node(name=s1_name)
s2_node = slice.get_node(name=s2_name)
tap_threads = []

for node, tap_network in tap_networks.items():
    host = slice.get_node(name=node)
    ip, switch_addr, port = tap_network
    print("Establishing TAP virtual connection between {} and {}...".format(node, 's1' if switch_addr == s1_addr else 's2')) 
    if switch_addr == s2_addr:
        ts = s2_node.execute_thread("sudo docker exec -w /root -d fabric_p4 python3 l2direct.py {} undefined".format(port))
    else:
        ts = s1_node.execute_thread("sudo docker exec -w /root -d fabric_p4 python3 l2direct.py {} undefined".format(port))
    time.sleep(1)
    tc = host.execute_thread("sudo docker exec -w /root -d fabric_p4 python3 l2direct.py {} {} {}".format(switch_addr, port, ip))
    time.sleep(1)
    tap_threads.append(ts)
    tap_threads.append(tc)
```

## Setup local network for s2

```python
s2_node = slice.get_node(name=s2_name)

# compile and install
# Note: 1 -> h3, 2 -> h4, 3 -> s1
# if the port mapping is changed, you need to change the command files in the next cell
s2_node.execute('sudo docker exec -w /workdir fabric_p4 p4c --p4runtime-files basic.txt --target bmv2 --arch v1model basic.p4')
s2_node.execute('sudo docker exec -w /workdir -d fabric_p4 simple_switch -i 1@tap0 -i 2@tap1 -i 3@tap2 basic.json')

print("Installed P4 rules on s2")
```

```python
h3_node = slice.get_node(name=d3_name)
h4_node = slice.get_node(name=d4_name)

extract_mac = "ip addr show dev tap0 | grep 'link/ether' | awk '{print $2}'"
(h3_mac, _) = h3_node.execute(extract_mac)
h3_mac = h3_mac.strip()
h3_port = 1
(h4_mac, _) = h4_node.execute(extract_mac)
h4_mac = h4_mac.strip()
h4_port = 2

commands_tpl = '''table_add ipv4_lpm MyIngress.ipv4_forward 10.0.1.10/32 => {h3_mac}
table_add ipv4_lpm MyIngress.ipv4_forward 10.0.1.11/32 => {h4_mac}
table_set_default ethernet_exact MyIngress.drop
table_add ethernet_exact MyIngress.broadcast FF:FF:FF:FF:FF:FF =>
table_add ethernet_exact MyIngress.ethernet_forward {h3_mac} => {h3_port}
table_add ethernet_exact MyIngress.ethernet_forward {h4_mac} => {h4_port}
mc_mgrp_create 1
mc_mgrp_create 2
mc_mgrp_create 3
mc_node_create 0 2 3
mc_node_create 1 1 3
mc_node_create 2 1 2
mc_node_associate 1 0
mc_node_associate 2 1
mc_node_associate 3 2
'''

commands = commands_tpl.format(h3_mac=h3_mac, h4_mac=h4_mac, h3_port=h3_port, h4_port=h4_port)
print(commands)

s2_node = slice.get_node(name=s2_name)
open("s2-commands.txt", 'w').write(commands)
s2_node.upload_file("s2-commands.txt", 's2-commands.txt')
s2_node.execute("sudo sh -c 'cat s2-commands.txt | docker exec -i fabric_p4 simple_switch_CLI'")
print("Added P4 table rules on s2")
```

```python
h1_node = slice.get_node(name=d1_name)
h2_node = slice.get_node(name=d2_name)
h5_node = slice.get_node(name=attacker_name)

extract_mac = "ip addr show dev tap0 | grep 'link/ether' | awk '{print $2}'"
(h1_mac, _) = h1_node.execute(extract_mac)
h1_mac = h1_mac.strip()
(h2_mac, _) = h2_node.execute(extract_mac)
h2_mac = h2_mac.strip()
(h5_mac, _) = h5_node.execute(extract_mac)
h5_mac = h5_mac.strip()

command_tpl = '''table_add ethernet_exact MyIngress.ethernet_forward {h1_mac} => 3
table_add ethernet_exact MyIngress.ethernet_forward {h2_mac} => 3
table_add ethernet_exact MyIngress.ethernet_forward {h5_mac} => 3
'''
command = command_tpl.format(h1_mac=h1_mac, h2_mac=h2_mac, h5_mac=h5_mac)
print(command)

s2_node = slice.get_node(name=s2_name)
open("s2-commands.txt", 'w').write(command)
s2_node.upload_file("s2-commands.txt", 's2-commands.txt')
s2_node.execute("sudo sh -c 'cat s2-commands.txt | docker exec -i fabric_p4 simple_switch_CLI'")
print("Added P4 table rules for hosts behind s1 on s2")
```
